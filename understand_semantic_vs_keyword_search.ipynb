```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
import re
```

```python
documents = [
    "This is a list which containig sample documents.",
    "Keywords are important for keyword-based search.",
    "Document analysis involves extracting keywords.",
    "Keyword-based search relies on sparse embeddings."
]
```

```python
query="what is keyword-based search?"
```

```python
# preprocess the documents
def preprocessing_text(text: str):
    processed_text = text.lower()
    # remove punctuation
    processed_text = re.sub(r'[^\w\s]', '', processed_text)
    return processed_text
```

```python
processed_documents = [preprocessing_text(doc) for doc in documents]
processed_documents
```

```python
processed_query = preprocessing_text(query)
processed_query
```

```python
# declare vectorizor
vectorizer = TfidfVectorizer()
```

```python
X = vectorizer.fit_transform(processed_documents)
```

```python
X.toarray()
```

```python
y = vectorizer.transform([processed_query])
y.toarray()
```

```python
similarities = cosine_similarity(X, y)
similarities
```

```python
ranked_indices=np.argsort(similarities,axis=0)[::-1].flatten()
ranked_indices
```

```python
ranked_documents = [documents[i] for i in ranked_indices]
ranked_documents
```

```python
# Output the ranked documents
for i, doc in enumerate(ranked_documents):
    print(f"Rank {i+1}: {doc}")
```
